```python
import os, string
import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from tabulate import tabulate
from collections import OrderedDict, Counter
```

```python
f = open('./src/finnish/finnish.txt')
raw = f.read()
f.close()

stripped = str(raw.split()).lower().translate(
    str.maketrans({'.':None})).translate(
    str.maketrans({'"':None})).translate(str.maketrans({',':None}))
out = stripped.translate(str.maketrans('', '', string.punctuation))

tokens = word_tokenize(out)
words = [w.lower() for w in tokens]
uniqueWords = list(OrderedDict.fromkeys(words))
```

```python
def wordTest(uniq):
    uwords = []
    ywords = []
    for i in uniq:
        if "u" in i:
            uwords.append(i)
        elif "y" in i:
            ywords.append(i)
    uwords = list(OrderedDict.fromkeys(words))
    return [uwords, ywords]

def vowelCounter(vowelsWords):
    vCount = []
    for k, v in vowelsWords:
        vCount.append(k)
    return Counter(vCount)

def vowelCheck(vlist):
    vowelsWords = []
    vowels = ["a", "e", "i", "o", "u", "ä", "ö"]
    for i_word in vlist:
        def hasVowel(inword):
            for vowel in vowels:
                if vowel in inword:
                    vowelsWords.append([vowel, inword])
            return(vowelsWords)
        hasVowel(i_word)
    return(vowelsWords)
```

```python
uwords = vowelCounter(vowelCheck(wordTest(uniqueWords)[0]))
ywords = vowelCounter(vowelCheck(wordTest(uniqueWords)[1]))
```
