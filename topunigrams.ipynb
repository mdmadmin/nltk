```python
from nltk import *
from nltk.tokenize import word_tokenize
import json
import string
import pandas as pd
from tabulate import tabulate
```

```python
file = './src/unigrams.txt'
```

```python
with open(file) as parseFile:
    lines = parseFile.read()
    cleanLines = lines.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(cleanLines)
    tokenDict = {tokens[i]: tokens[i + 1] for i in range(0, len(tokens), 2)}
```

```python
df = pd.DataFrame.from_dict(tokenDict, orient='index', columns=['Count'])
pdtabulate=lambda df:tabulate(df,headers=['ngram', 'count'],tablefmt='pretty', disable_numparse=True, colalign=("left","left"))
print(pdtabulate(df))
```
